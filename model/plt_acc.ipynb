```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
import seaborn as sns
```

```python
fname = 'trained_model/VGG_binary_new_augmentation_epoch_accuracy.csv'
df = pd.read_csv(fname)
df
```

```python
# plot loss
x = df['epoch']
loss = df['training_loss']
plt.figure(figsize=(5, 4))
plt.plot(x, loss, '-o', label='Training set')
# plt.xlim(0, 100)
plt.xlim(0, 40)
# plt.ylim(0.3, 0.7)
plt.ylim(0.3, 0.6)
plt.xlabel('Epoch', fontname='Arial', fontsize=13)
plt.ylabel('Loss', fontname='Arial', fontsize=13)
# plt.xticks(np.arange(0, 101, 20), fontname='Arial', fontsize=12)
plt.xticks(np.arange(0, 41, 10), fontname='Arial', fontsize=12)
# plt.yticks(np.arange(0.3, 0.71, 0.1), fontname='Arial', fontsize=12)
plt.yticks(np.arange(0.3, 0.61, 0.1), fontname='Arial', fontsize=12)
plt.title('Training Loss', fontname='Arial', fontsize=13)
ax = plt.gca()
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['top'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.tick_params(axis="x", length=8, width=1.5, direction="in")
ax.tick_params(axis="y", length=8, width=1.5, direction="in")
ax.tick_params(which='minor', length=4, width=1.5, direction="in")
# plt.legend(loc='lower right')
plt.tight_layout()
```

```python
# plot accuracy.
x = df['epoch']
train_acc = df['training_accuracy']
test_acc = df['test_accuracy']
plt.figure(figsize=(5, 4))
plt.plot(x, train_acc*100, '-o', color='tab:blue', label='Training set')
plt.plot(x, test_acc*100, '-o', color='tab:red', label='Test set')
# plt.axhline(y=100, color='r', linestyle='--')
# plt.xlim(0, 100)
# plt.ylim(50, 100)
plt.xlim(0, 41)
plt.ylim(70, 100)
plt.xlabel('Epoch', fontname='Arial', fontsize=13)
plt.ylabel('Accuracy ($\%$)', fontname='Arial', fontsize=13)
# plt.xticks(np.arange(0, 101, 20), fontname='Arial', fontsize=12)
# plt.yticks(np.arange(50, 101, 10), fontname='Arial', fontsize=12)
plt.xticks(np.arange(0, 41, 10), fontname='Arial', fontsize=12)
plt.yticks(np.arange(70, 101, 10), fontname='Arial', fontsize=12)
plt.title('Model Accuracy', fontname='Arial', fontsize=13)
ax = plt.gca()
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['top'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.tick_params(axis="x", length=8, width=1.5, direction="in")
ax.tick_params(axis="y", length=8, width=1.5, direction="in")
ax.tick_params(which='minor', length=4, width=1.5, direction="in")
plt.legend(loc='lower right', prop={'family': 'Arial', 'size': 12})
plt.tight_layout()
plt.savefig('trained_model/VGG_binary_new_augmentation_epoch_accuracy.png')
plt.savefig('trained_model/VGG_binary_new_augmentation_epoch_accuracy.eps')
```
